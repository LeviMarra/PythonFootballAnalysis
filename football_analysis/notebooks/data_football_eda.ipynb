```python
from src.data_loader import load_csv


player_profiles = load_csv("player_profiles.csv")
market_values = load_csv("player_market_value.csv")
player_performances = load_csv("player_performances.csv")
player_injuries = load_csv("player_injuries.csv")


# First Analysis
print("üìä FOOTBALL DATA EXPLORATION")
print("=" * 50)

datasets = {
    'Player Profiles': player_profiles,
    'Market Values': market_values,
    'Player Performances': player_performances,
    'Player Injuries': player_injuries
}

for name, df in datasets.items():
    print(f"\n{name}:")
    print(f"  Shape: {df.shape}")
    print(f"  Columns: {list(df.columns)}")
    print(f"  Sample:")
    display(df.head(5))
```

```python
# C√©lula 4 - FILTRAR JOGADORES BRASILEIROS
print("üáßüá∑ AN√ÅLISE DE JOGADORES BRASILEIROS")
print("=" * 45)

# Filtrar brasileiros
brazilian_players = player_profiles[player_profiles['country_of_birth'] == 'Brazil']
print(f"Total de jogadores brasileiros: {len(brazilian_players)}")

# Top posi√ß√µes
print("\nüìä POSI√á√ïES MAIS COMUNS:")
print(brazilian_players['position'].value_counts().head(10))

# Clubes atuais
print("\nüèÜ CLUBES ATUAIS (TOP 10):")
print(brazilian_players['current_club_name'].value_counts().head(10))
```

```python
# C√©lula 5 - AN√ÅLISE DE VALORES
print("üí∞ AN√ÅLISE DE VALORES DE MERCADO")
print("=" * 40)

# Juntar valores com perfis
player_values = market_values.merge(player_profiles[['player_id', 'player_name', 'country_of_birth']], on='player_id')

# Valores mais altos
top_values = player_values.sort_values('value', ascending=False).head(10)
print("TOP 10 VALORES DE MERCADO:")
print(top_values[['player_name', 'country_of_birth', 'value']])
```

```python
# C√©lula 7 - DESEMPENHO DE BRASILEIROS
print("‚öΩ DESEMPENHO DE JOGADORES BRASILEIROS")
print("=" * 48)

# Juntar desempenho com jogadores BR
br_performances = player_performances.merge(brazilian_players[['player_id']], on='player_id')

# Top artilheiros
top_scorers = br_performances.groupby('player_id')['goals'].sum().sort_values(ascending=False).head(10)
top_scorers_names = top_scorers.reset_index().merge(player_profiles[['player_id', 'player_name']], on='player_id')
print("üéØ TOP ARTILHEIROS BRASILEIROS:")
print(top_scorers_names[['player_name', 'goals']])
```
